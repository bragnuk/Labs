1. DATASET iris
1.1 3D-графік

```python
df = px.data.iris()

fig1 = px.scatter_3d(
    df,
    x='sepal_length',
    y='sepal_width',
    z='petal_length',
    color='species',
    title='3D розподіл Iris'
)

fig1.update_layout(title_x=0.5)
fig1.show()
```

1.2 Гістограма petal_width

```python
fig2 = px.histogram(
    df,
    x='petal_width',
    color='species',
    barmode='overlay',
    title='Розподіл petal_width'
)

fig2.update_layout(title_x=0.5)
fig2.show()
```

1.3 Лінійний графік середнього sepal_length

```python
df_grouped = df.groupby(['species', 'sepal_width'], as_index=False)['sepal_length'].mean()

fig3 = px.line(
    df_grouped,
    x='sepal_width',
    y='sepal_length',
    color='species',
    title='Залежність sepal_length від sepal_width'
)

fig3.update_layout(title_x=0.5)
fig3.show()
```

2. DATASET gapminder
2.1 Гістограма ВВП (2007, топ 10)

```python
df = px.data.gapminder()
df_2007 = df[df.year == 2007].nlargest(10, 'gdpPercap')

fig4 = px.bar(
    df_2007,
    x='country',
    y='gdpPercap',
    color='gdpPercap',
    title='Топ 10 країн за ВВП (2007)'
)

fig4.update_layout(title_x=0.5)
fig4.show()
```

2.2 Кругова діаграма Європа

```python
df_europe = df[(df.year == 2007) & (df.continent == 'Europe')].nlargest(5, 'gdpPercap')

fig5 = px.pie(
    df_europe,
    names='country',
    values='gdpPercap',
    title='Топ 5 країн Європи за ВВП'
)

fig5.update_layout(title_x=0.5)
fig5.show()
```

2.3 Динаміка населення

```python
df_filtered = df[df.country.isin(['Canada', 'Australia', 'New Zealand'])]

fig6 = px.line(
    df_filtered,
    x='year',
    y='pop',
    color='country',
    title='Динаміка населення'
)

fig6.update_layout(title_x=0.5)
fig6.show()
```

3. Тригонометричні функції
3.1 y = 2cos(x − π/4), [0; 5π]

```python
x = np.linspace(0, 5*np.pi, 500)
y = 2 * np.cos(x - np.pi/4)

fig7 = go.Figure()
fig7.add_trace(go.Scatter(x=x, y=y, mode='lines'))

fig7.update_layout(title='y = 2cos(x − π/4)', title_x=0.5)
fig7.show()
```

3.2 y = 1/2 sin(2x + π/3), [0; 2π]

```python
x = np.linspace(0, 2*np.pi, 500)
y = 0.5 * np.sin(2*x + np.pi/3)

fig8 = go.Figure()
fig8.add_trace(go.Scatter(x=x, y=y, mode='lines'))

fig8.update_layout(title='y = 1/2 sin(2x + π/3)', title_x=0.5)
fig8.show()
```

3.3 y = 2tg(1/2 x + π/6), [-2π; 2π]

```python
x = np.linspace(-2*np.pi, 2*np.pi, 500)
y = 2 * np.tan(0.5*x + np.pi/6)

fig9 = go.Figure()
fig9.add_trace(go.Scatter(x=x, y=y, mode='lines'))

fig9.update_layout(title='y = 2tg(1/2 x + π/6)', title_x=0.5)
fig9.show()
```
